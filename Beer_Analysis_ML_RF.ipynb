```python
# Stratified Sampling ?? stratified based on ratio?
beer_styles = beer_reviews_complete_df['style'].unique().tolist()
sample_size_per_style = 11100
df_stratified = beer_reviews_complete_df[beer_reviews_complete_df["style"] == "Does not exist"]

for styles in beer_styles:
    df_sample  = beer_reviews_complete_df[beer_reviews_complete_df["style"] == styles]
    df_sample = df_sample.sample(n = sample_size_per_style)
    df_stratified = df_sample.append(df_stratified)

df_stratified
```

```python
#reset index
df_stratified.reset_index(inplace=True)
df_stratified
```

```python
df_stratified.state_beer.value_counts()
```

```python
df_stratified.city.value_counts()
```

```python
df_stratified.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_reviews_encode = df_stratified.copy()
df_reviews_encode['availability'] = le.fit_transform(df_reviews_encode['availability'])
df_reviews_encode['state_beer'] = le.fit_transform(df_reviews_encode['state_beer'])
```

```python
df_reviews_encode.head()
```

```python
# imports for random forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
#define the features set
X = df_reviews_encode.copy()
X = X.drop(columns =["index","beer_id","beer_name","brewery_id","country_beer","brewery_name","city","style"])
X
```

```python
# Define the target set
y = df_reviews_encode["style"].ravel()
y
```

```python
# Splitting into Train and Test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=100)
```

```python
# Create a StandardScaler instance
scaler = StandardScaler()
# Fitting the Standard Scaler with the training data.
X_scaler = scaler.fit(X_train)

# Scaling the data.
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Create a random forest classifier
rf_model = RandomForestClassifier(n_estimators = 128, random_state = 80)
```

```python
# fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train)
```

```python
# making predictions using testing data
predictions = rf_model.predict(X_test_scaled)
```

```python
# Calculating the confusion matrix
cm = confusion_matrix(y_test, predictions)
cm_df = pd.DataFrame(cm)
```

```python
# calculate the accuracy score
acc_score = accuracy_score(y_test, predictions)
```

```python
# Display results
print("Confusion Matrix")
display(cm_df)
print(f"ACcuracy Score: {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```

```python
# calculate feature importance in the Random Forest model
importances = rf_model.feature_importances_
importances
```

```python
# Sort the features by importances
sorted(zip(rf_model.feature_importances_, X.columns), reverse = True)
```

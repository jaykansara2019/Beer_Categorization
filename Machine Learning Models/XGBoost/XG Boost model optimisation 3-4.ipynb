```python
#Find outliers
def find_outliers_IQR(df_stratified):
    q1=df_stratified.quantile(0.25)
    q3=df_stratified.quantile(0.75)
    IQR=q3-q1
    outliers = df_stratified[((df_stratified<(q1-1.5*IQR)) | (df_stratified>(q3+1.5*IQR)))]
    return outliers
```

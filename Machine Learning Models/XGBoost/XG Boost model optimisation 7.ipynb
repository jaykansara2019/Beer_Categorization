```python
# Import dependencies 
import pandas as pd
import os
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
# Creating dataframe for beers and filtering for the US
encoded_df3 = pd.read_csv('encoded_df3.csv')
encoded_df3
```

```python
#Encode string values
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
encoded_df3 = encoded_df3.copy()
encoded_df3 ['style'] = le.fit_transform(encoded_df3 ['style'])

encoded_df3
```

```python
#Create a seperate df for y
style_df = encoded_df3.filter(['style'], axis=1)
style_df
```

```python
#Drop y from the main df
encoded_df3 = encoded_df3.drop(['style'], axis=1)
encoded_df3
```

```python
#Check the data types for each culumn
encoded_df3.dtypes
```

```python
#Scale the data
from sklearn.preprocessing import StandardScaler
data_scaler = StandardScaler()
```

```python
scaled_df = data_scaler.fit_transform(encoded_df3)
scaled_df
```

```python
# Define features set
X = scaled_df.copy()
X
```

```python
# Define target vector
y = style_df.values
```

```python
from numpy import loadtxt
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
# split data into train and test sets

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)
```

```python
# fit model no training data
model = XGBClassifier()
model.fit(X_train, y_train)
```

```python
# make predictions for test data
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred]
```

```python
#calculate matrix and accuracy score
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
cm = confusion_matrix(y_test, predictions)
cm_df = pd.DataFrame(cm)
acc_score = accuracy_score(y_test, predictions)
```

```python
# evaluate predictions
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
print("Confusion Matrix")
print("Classification Report")
print(classification_report(y_test, predictions))
```

```python
# Stratified Sampling
beer_styles = beer_reviews_complete_df['style'].unique().tolist()
sample_size_per_style = 11100
df_stratified = beer_reviews_complete_df[beer_reviews_complete_df["style"] == "Does not exist"]

for styles in beer_styles:
    df_sample  = beer_reviews_complete_df[beer_reviews_complete_df["style"] == styles]
    df_sample = df_sample.sample(n = sample_size_per_style)
    df_stratified = df_sample.append(df_stratified)

df_stratified
```
